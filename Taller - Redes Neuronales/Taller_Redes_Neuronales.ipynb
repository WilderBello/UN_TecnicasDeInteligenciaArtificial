```python
import numpy as np
```

```python
# Numero de elementos en el dataset
N = 10
x_i = list(range(1,N+1))
#print(x_i)
# 1. Inicializar aleatoriamente los parametros(pesos) w_i
W = np.random.randint(N, size=(N, N))
#print(W)
# 2. Definir el numero maximo de iteraciones
k = 2
def FA(connection):
    FA = list()
    for con in connection:
        if con < 0:
            FA.append(0)
        else:
            FA.append(1)
    return FA
error = 1
while (error != 0 or k == 0):
    # 3. Calcular la suma net
    net = list()
    i = 0
    for w_i in W:
        suma = 0
        for j in w_i:
            suma += j*x_i[i]
        net.append(suma)
        i += 1
    # 4. Calcule el calor de salida Y
    y = FA(net)
    # 5. Calcule el error y_barra - y
    errores = list()
    i = 0
    for y_element in y:
        y_barra = 0
        error = y_barra - y_element
        errores.append(error)
    
    for error in errores:
        if error != 0:
            W[k+1] = W[k] + error*x_i[i]
            print(W)
        else: 
            i += 1
    if i == len(errores):
        break
    k -= 1
```
