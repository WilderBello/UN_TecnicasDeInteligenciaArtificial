# Proyecto de reconocimiento de Ojos

# 0. Banco de imagenes

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

imagen = 'Images/1.jpg'
img = cv2.imread(imagen)
```

# 1. Histograma


```python
cv2.imshow('Imagen.jpg', img)

color = ('b', 'g', 'r')

for i, c in enumerate(color):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = c)
    plt.xlim([0, 256])

plt.show()
cv2.waitKey() 

cv2.destroyAllWindows()
```

```python
imagen_gris = cv2.imread(imagen, cv2.IMREAD_GRAYSCALE)
imagen_gris = cv2.equalizeHist(imagen_gris)

cv2.imshow('Histogramas', imagen_gris)
cv2.waitKey() 
```

# 2. Bordes

```python
# Pasar imagen a escala de grises.
#escalaGrises = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
escalaGrises = cv2.imread(imagen, cv2.IMREAD_GRAYSCALE)
escalaGrises = cv2.equalizeHist(escalaGrises)
# Histograma para detectar la pupila
hist = cv2.calcHist(escalaGrises, [0], None, [256], [0, 256])
plt.plot(hist, color = 'gray')
plt.xlabel('intensidad de iluminacion')
plt.ylabel('cantidad de pixeles')
plt.show()
# Suavizado de imagen - filtro mediana
blur = cv2.medianBlur(escalaGrises, 5)
# Detector de bordes Canny
bordes = cv2.Canny(blur, 30, 150)

cv2.imshow('Imagen', bordes)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
