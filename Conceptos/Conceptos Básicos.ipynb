# Relaciones básicas:
1. Vecindad: 
Serian los vecinos vistos en una ubicación como panel de abeja, de tal manera que se vea el punto central.
https://www.pinterest.es/pin/843439836457212625/
De tal manera que se tengan los puntos cercanos al pixel en vertical, horizontal y diagonal. Iniciando por el mas cercano en diagonal superior derecha y siguiendo en contra de las manecillas del reloj.

- El punto (0,0) es la esquina superior izquierda de la imagen

2. Conectividad:
Se debe mirar si los vecinos tienen el mismo valor del pixel que se está revisando. Se establece un criterio de conectividad con los parámetros base para comprobar la conectividad y posteriormente se busca la <strong>M conectividad</strong> que al ver las regiones mediante la conexión, se puede notar la conexión entre las mismas.

3. Camino:
Secuencia (S) de puntos adyacentes que forman el camino desde el punto (0,0) hasta (n,n). 

4. Métrica de distancia:
D es una métrica de distancia si cumple con las propiedades definidas:

- D(p,q) >= 0, cuando p=q es 0. Semidefinida positiva.

- Simétrica D(p,q) = D(q,p)

- Desigualdad Triangular.

- Medidas de distancia: Euclídea $D_0$, $D_4$ que realiza movimientos en (x,y), $D_8$ que realiza movimientos diagonales.

5. Histograma:

- Distribución de grises en una imagen
- Función de densidad de probabilidad de niveles de gris (medida: nivel de gris, # veces: del nivel de gris)
    - Abcisas: Rango posible de densidades de grises.
    - Ordenadas: Número de píxeles para cada nivel de gris.

Un histograma se puede ver como un arreglo ordenado de acuerdo el nivel de gris con la posición dada por el número de veces que se repite.

- En OpenCV: cv2.calcHist(parámetros), .shape muestra el tamaño del objeto, plt.plot(histograma) para poder plotearlo.
